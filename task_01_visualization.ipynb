```python
import pandas as pd

age_data = {
    "Age": [18, 22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 23, 27, 32, 38, 44, 48, 52, 58]
}

df = pd.DataFrame(age_data)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=8, kde=True)

plt.title("Age Distribution of Population")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

### Observations

- The histogram shows the distribution of ages in the dataset.
- Most individuals fall in the age range of 20 to 50 years.
- The frequency decreases as age increases beyond 50.
- The KDE curve indicates a slightly right-skewed distribution.
